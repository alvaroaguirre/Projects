# Benford's Law

I learnt about Benford's Law a couple of days ago. It says that, as strange as it sounds, the frequency distribution of leading digits in many real-life sets of numerical data is non-uniform. Not only it is non-uniform, but it actually follows this distribution:

<img src="Benfords-Law.jpg" width="200" height="120"> 

I couldn't believe it. This is supposed to hold for any type of real-life data, regardless of the unit. Sounds absurd, but it is widely used in different fields, from detecting deep fakes to cooked accounting books. So I decided to do a little test using two databases, one with financial data, and another with data on daily Coronavirus cases.

## Daily stock prices - 2019

I downloaded the whole CRSP database for 2019, which includes the daily prices for all stocks from 2019. Let's see what we find:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("crsp.csv")
```

We have almost 2 million columns:

```python
np.shape(df)
```

I want to do a little bit of cleaning. Apart from removing NAs, I want to remove negative values. CRSP can fill up prices with the negative of the previous day price if it doesn't have a value for that day:

```python
# Removing NAs
df = df.dropna(subset=['PRC'])

# Keeping only positive values
df = df[df.PRC > 0]
```

Let's see what the average price is:

```python
avg_price = np.mean(df.PRC)
std = np.std(df.PRC)

print(" Average price:", round(avg_price,2), "USD \n",
     "Standard deviation:", round(std,2))
```

Now let's use the `apply` method of pandas to create a new `Benford` column, which contains the leading digit of the price. We have to do a little trick here, since prices can be very small, so have multiple leading zeros, I will use `replace()` to only keep non-zeros:

```python
df['Benford'] = df.apply(lambda row: int(str(row.PRC).replace('0','').replace('.','')[0]), axis = 1)
```

```python
benfords_law = round((df.Benford.value_counts() / len(df.Benford))*100,2)
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = benfords_law.plot.bar(rot = 0, title = "Benford's Law for 2019 stock prices");
ax.set_xlabel("Leading non-zero digit in the stock price")
ax.set_ylabel("Distribution")
for p in ax.patches:
    ax.annotate(str(p.get_height()) + "%", (p.get_x() * 1.01, p.get_height() + 0.1))
fig.savefig('benford_stocks.png', bbox_inches="tight")
```

## Coronavirus

I am using a dataset downloaded from *Our World in Data*, which includes a lot of information on Covid 19 by day and by country: https://ourworldindata.org/coronavirus-source-data

```python
df = pd.read_csv("owid-covid-data.csv")
df.head()
```

We have over 35k observations:

```python
np.shape(df)
```

Let's remove NAs and zeros:

```python
# Removing NAs
df = df.dropna(subset=['new_cases'])

# Keeping only positive values
df = df[df.new_cases > 0]
```

Now let's create a `Benford` column using the `apply` pandas method to extract the leading non-zero digit. 

```python
df['Benford'] = df.apply(lambda row: int(str(row.new_cases).replace('0','').replace('.','')[0]), axis = 1)
```

Let's extract the results in another dataframe and format them:

```python
benfords_law = round((df.Benford.value_counts() / len(df.Benford))*100,2)
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = benfords_law.plot.bar(rot = 0, title = "Benford's Law for daily new COVID-19 cases");
ax.set_xlabel("Leading non-zero digit")
ax.set_ylabel("Distribution")
for p in ax.patches:
    ax.annotate(str(p.get_height()) + "%", (p.get_x() * 1.01, p.get_height() + 0.1))
fig.savefig('benfords_covid.png', bbox_inches="tight")
```

## Changing units

Benford's law is supposed to hold regardless of the units. So now I'll repeat the same exercise using new daily cases per million, instead of absolute number of new cases. Let's see what happens:

```python
df = pd.read_csv("owid-covid-data.csv")
# Removing NAs
df = df.dropna(subset=['new_cases_per_million'])

# Keeping only positive values
df = df[df.new_cases_per_million > 0]
```

```python
df['Benford_millions'] = df.apply(lambda row: int(str(row.new_cases_per_million).replace('0','').replace('.','')[0]), axis = 1)
```

Let's extract the results in another dataframe and format them:

```python
benfords_law_million = round((df.Benford_millions.value_counts() / len(df.Benford_millions))*100,2)
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = benfords_law_million.plot.bar(rot = 0, title = "Benford's Law for daily new COVID-19 cases per million");
ax.set_xlabel("Leading non-zero digit")
ax.set_ylabel("Distribution")
for p in ax.patches:
    ax.annotate(str(p.get_height()) + "%", (p.get_x() * 1.01, p.get_height() + 0.1))
fig.savefig('benfords_covid_per_million.png', bbox_inches="tight")
```

Still holds! Amazing!
