```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
data = pd.read_csv(url, error_bad_lines=False)
```

```python
df = data.loc[data['iso_code'].isin(['ESP','GBR','PER'])]
df['date'] = pd.to_datetime(df['date'])
df.index = pd.DatetimeIndex(df.pop('date'))
df = df.sort_index().last('30D')
```

```python
df.loc[df['iso_code']=='ESP'].plot(y = 'new_cases_smoothed', label = '7-day rolling average');
plt.title('New cases ESP - Last 30 days');
```
